```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define true values (y_true) and model predictions (y_pred)
y_true = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])  # One-hot encoded true labels
y_pred_1 = np.array([[0.1, 0.8, 0.1], [0.7, 0.2, 0.1], [0.2, 0.1, 0.7]])  # First prediction set
y_pred_2 = np.array([[0.2, 0.7, 0.1], [0.6, 0.3, 0.1], [0.3, 0.1, 0.6]])  # Slightly modified predictions

# Step 2: Compute Mean Squared Error (MSE) and Categorical Cross-Entropy (CCE)
mse_loss_fn = tf.keras.losses.MeanSquaredError()
cce_loss_fn = tf.keras.losses.CategoricalCrossentropy()

mse_1 = mse_loss_fn(y_true, y_pred_1).numpy()
cce_1 = cce_loss_fn(y_true, y_pred_1).numpy()

mse_2 = mse_loss_fn(y_true, y_pred_2).numpy()
cce_2 = cce_loss_fn(y_true, y_pred_2).numpy()

# Step 3: Print loss values for different predictions
print(f"Loss Values:")
print(f"Set 1 - MSE: {mse_1:.4f}, CCE: {cce_1:.4f}")
print(f"Set 2 - MSE: {mse_2:.4f}, CCE: {cce_2:.4f}")

# Step 4: Plot bar chart comparing losses
labels = ['MSE (Set 1)', 'CCE (Set 1)', 'MSE (Set 2)', 'CCE (Set 2)']
values = [mse_1, cce_1, mse_2, cce_2]

plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['blue', 'red', 'blue', 'red'], alpha=0.7)
plt.xlabel("Loss Type")
plt.ylabel("Loss Value")
plt.title("Comparison of MSE and Categorical Cross-Entropy Loss")
plt.show()
```
